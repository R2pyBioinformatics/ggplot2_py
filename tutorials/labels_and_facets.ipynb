# Labels, Titles, and Facet Strips

This notebook validates three Grammar of Graphics features:
1. **Axis titles** — `labs(x=..., y=...)`
2. **Plot title / subtitle / caption** — `labs(title=..., subtitle=..., caption=...)`
3. **Facet strip labels** — column and row annotations in `facet_grid` / `facet_wrap`

```python
from ggplot2_py import *
from ggplot2_py.plot import GGPlot
from ggplot2_py.datasets import mpg

GGPlot.fig_width = 7
GGPlot.fig_height = 5
GGPlot.fig_dpi = 150
```

## 1. Axis Titles

`labs(x=..., y=...)` adds descriptive axis titles below and to the left of the plot.

```python
ggplot(mpg, aes('displ', 'hwy')) + geom_point() + labs(x='Engine Displacement (L)', y='Highway MPG')
```

## 2. Plot Title, Subtitle, and Caption

Titles appear at the top, captions at the bottom-right.

```python
(
    ggplot(mpg, aes('cty', 'hwy', colour='class'))
    + geom_point()
    + labs(
        title='Fuel Economy',
        subtitle='EPA data for 234 popular car models (1999-2008)',
        x='City MPG',
        y='Highway MPG',
        caption='Source: fueleconomy.gov',
    )
)
```

## 3. Facet Strip Labels

`facet_grid` adds strip labels above each column and to the right of each row.

```python
# facet_grid with both row and column strips
(
    ggplot(mpg, aes('cty', 'hwy'))
    + geom_point()
    + facet_grid('year', 'drv')
)
```

```python
# facet_wrap strips
(
    ggplot(mpg, aes('cty', 'hwy'))
    + geom_point()
    + facet_wrap('class')
)
```

## 4. All Combined

Title + subtitle + axis labels + facet strips working together.

```python
(
    ggplot(mpg, aes('cty', 'hwy'))
    + geom_point(aes(colour='displ'))
    + geom_smooth(method='lm', linewidth=0.3)
    + scale_colour_viridis_c()
    + facet_grid('year', 'drv')
    + labs(
        title='Fuel Economy by Year and Drive Train',
        subtitle='Linear trend with confidence band',
        x='City MPG',
        y='Highway MPG',
        caption='Source: mpg dataset',
    )
    + theme_minimal()
    + theme(panel_grid_minor=element_blank())
)
```

```python
# Single-panel with all labels
(
    ggplot(mpg, aes('displ', 'hwy'))
    + geom_point(aes(colour='class'), size=2)
    + geom_smooth(method='lm')
    + labs(
        title='Highway Fuel Efficiency',
        subtitle='Larger engines tend to be less efficient',
        x='Engine Displacement (litres)',
        y='Highway Miles per Gallon',
        caption='Data: EPA fuel economy',
    )
)
```
